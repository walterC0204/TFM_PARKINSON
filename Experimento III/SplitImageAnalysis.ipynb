# **Clase SplitImageAnalysis**

```python
class SplitImageAnalysis:

  def __init__(self, test_type, t_group):
    """
    @test_type: tipo de prueba -> Meander o Spiral
    @t_group: tipo de grupo de imagen -> Patients o Controls
    """
    self.mark = 'white'
    self.p_mask = 0.2
    self.path_load = f'/content/drive/MyDrive/{test_type}{t_group}_cutImg'

  def check_pixel(self, image):
    """
    Def: Verifica si la imagen es completamente blanca
    @image: imagen a analizar
    @return: True si la imagen es completamente blanca, False en caso contrario
    """
    flag = False
    n_all_pixel = image.shape[0] * image.shape[1]
    num_pixel = np.sum(image == 255) #
    if num_pixel == n_all_pixel:
      flag = True
      #print(n_all_pixel)
      #print(num_pixel)
    return flag

  def run(self):
    """
    Def: Ejecuta el analisis de la imagen y renombra la imagen si es completamente blanca
    @return: None
    """
    list_folders = glob.glob(f'{self.path_load}/*')
    for folder in list_folders:
      for image_part in glob.glob(f'{folder}/*'):
         image = cv2.imread(image_part, 0)
         flag = self.check_pixel(image)
         if flag:
          os.rename(image_part, image_part.replace('.jpg', '_white.jpg'))
```
