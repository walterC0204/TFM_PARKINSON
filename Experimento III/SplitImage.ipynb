```python
import cv2
import glob
import os
```

# **Clase SplitImage**

```python
class SplitImage:

  def __init__(self, test_type, t_group):
    """
    @param test_type: Meander or Spiral
    @param t_group: Patients or Control
    @return: None
    """
    self.n_item = 5
    self.root_path = f'/content/drive/MyDrive/{test_type}{t_group}_cutImg'
    self.path_load = f'/content/drive/MyDrive/{test_type}{t_group}_out_resize'

  def create_folder(self, path):
    """
    @param path: path to create
    @return: None
    """
    dir = os.path.dirname(path+'/')
    if not os.path.exists(dir):
      os.makedirs(dir)
      print(f"Created folder: {dir}")
    # The line below caused the error as 'dir' is a string and does not have a .mkdir() method.
    # dir.mkdir(path)

  def cut_image(self, image, output_dir, name):
    """
    @param image: image to cut
    @param output_dir: path to save the images
    @param name: name of the image
    @return: None
    """
    self.create_folder(output_dir)
    image = cv2.imread(image)
    dim = image.shape
    #image_height, image_width, _ = image.shape
    sub_image_h = dim[0] // self.n_item
    sub_image_w = dim[1] // self.n_item
    #print("Init cut image ...")
    for row in range(self.n_item):
      for col in range(self.n_item):
        start_x = col * sub_image_w
        start_y = row * sub_image_h
        end_x = (col + 1) * sub_image_w
        end_y = (row + 1) * sub_image_h

        sub_image = image[start_y:end_y, start_x:end_x]
        output_path = os.path.join(output_dir, f"{name}_part_{row}_{col}.jpg")
        cv2.imwrite(output_path, sub_image)
        #print(f"Saved: {output_path}")

  def run(self):
    """
    @return: None
    """
    for image in glob.glob(f'{self.path_load}/*'):
      print(f" * Processing: {image}")
      name = image.split('/')[-1].split('.')[0]
      output_dir = os.path.join(self.root_path, name)
      self.cut_image(image, output_dir, name)
```

# Ejecución

```python
list_data = [('Meander','Patients'), ('Meander','Control'),
            ('Spiral','Patients'), ('Spiral','Control') ]

for data in list_data:
  print(f"Tipo exec: {data[0], data[1]}")
  c_image = SplitImage(data[0], data[1])
  c_image.run()
```
